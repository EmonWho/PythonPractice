Missing Value

```python
features_with_null_value=[features for features in train.columns if train[features].isnull().sum()>1]
for feature in features_with_null_value:
    print(feature, np.round(train[feature].isnull().mean(), 3),  ' % missing values')
```

There is a lot of missing value, So we need to find the Relationship between Missing value with sales price.

here Lets draw some graph to view 

```python
for feature in features_with_null_value:
    copy_train = train.copy()
    #Lets put a variable where 1 indicates observation is missing and 0 is available
    copy_train[feature]=np.where(copy_train[feature].isnull(), 1, 0)
    #calculate the mean price where information is present or missing
    copy_train.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

Numeric Value

```python
numeric_features=[feature for feature in train.columns if train[feature].dtypes != 'O']
print("Number of numeric value is ",len(numeric_features))
train[numeric_features].head()
```

```python
#have to find out how many number of Year related feature is available here
year_features=[feature for feature in numeric_features if 'Year' in feature or 'Yr' in feature]
year_features
```

```python
for feature in year_features:
    print(feature, train[feature].unique())
```

```python

## We will check whether there is a relation between year the house is sold and the sales price

train.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('The Median House Price')
plt.title("House Price vs YearSold")
```

```python
year_features
```

```python
##now analyze all year feature with salePrice

for feature in year_features:
    if feature!='YrSold':
        data=train.copy()
        ## We will capture the difference between year variable and year the house was sold for
        data[feature]=data['YrSold']-data[feature]

        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
#I will now analyze the discrete feature from this dataset
discrete_feature=[feature for feature in numeric_features if len(train[feature].unique())<30 and feature not in year_features+['Id']]
print("Discrete Variables Count is: {}".format(len(discrete_feature)))
```

```python
discrete_feature
```

```python
train[discrete_feature].head()
```

```python
#the relationship betwqeen then and sales price we try to find

for feature in discrete_feature:
  c_train=train.copy()
  c_train.groupby(feature)['SalePrice'].median().plot.bar()
  plt.xlabel("SalePrice")
  plt.ylabel(feature)    
  plt.show()
```

Continous Variable

```python

continous_feature=[feature for feature in numeric_features if feature not in discrete_feature+year_features+['Id']]
print("Continuous feature Count {}".format(len(continous_feature)))
```

```python
for feature in continous_feature:
    any=train.copy()
    any[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()    
```

```python
#Now we are working with logarithmic transformation
for feature in continous_feature:
  data=train.copy()
  if 0 in data[feature].unique():
    pass
  else:
    data[feature]=np.log(data[feature])
    data['SalePrice']=np.log(data['SalePrice'])
    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel("SalesPrice")
    plt.title(feature)
    plt.show()

 
   


           

 
```

Outliers

```python
for feature in continous_feature:
    data=train.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
        
    
```

```python
#Categorical Features
categorical_feature= [feature for feature in train.columns if train[feature].dtypes=='O']
print("Total Categorical feature is ",len(categorical_feature))
train[categorical_feature].head()
```

```python
for feature in categorical_feature:
  print("The feature is {} and number of unique value is {}".format(feature,len(train[feature].unique())))
```
